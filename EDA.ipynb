### Outliers

##### 1. Based on Box plots

```python
# filter the numeric variables from the data
df_num = df.select_dtypes(include = np.number)
df_num.head(5)
```

```python
Interpretation:we selected datatype as numerical to get only the numerical values. and we are checking the column head  
```

```python
# plot the boxplot for each variable
# subplots(): plot subplots
# figsize(): set the figure size
fig, ax = plot.subplots(2, 3, figsize=(15, 8))

# plot the boxplot using boxplot() from seaborn
# z: let the variable z define the boxplot
# x: data for which the boxplot is to be plotted
# orient: "h" specifies horizontal boxplot (for vertical boxplots use "v")
# whis: proportion of the IQR past the low and high quartiles to extend the plot whiskers
# ax: specifies the axes object to draw the plot onto
# set_xlabel(): set the x-axis label
# fontsize: sets the font size of the x-axis label
for variable, subplot in zip(df_num.columns, ax.flatten()):
    z = sns.boxplot(x = df_num[variable], orient = "h",whis=1.5 , ax=subplot) # plot the boxplot
    z.set_xlabel(variable, fontsize = 20)                                     # set the x-axis label
```

Interpretation:as we can observe Fuel Consumption Hwy (L/100 km),Fuel Consumption City (L/100 km),Fuel Consumption Comb (L/100 km) have any outliers which can influence our analysis.

##### 2. Based on IQR

```python
# obtain the first quartile
Q1 = df_num.quantile(0.25)

# obtain the third quartile
Q3 = df_num.quantile(0.75)

# obtain the IQR
IQR = Q3 - Q1

# print the IQR
print(IQR)
```

Interpretation:calculating IQR can be used to remove outlier. anything +/- 3 IQR can be considered as outlier

```python
# filter out the outlier values
# ~ : selects all rows which do not satisfy the condition
# any() : returns whether any element is True over the columns
# axis : "1" indicates columns should be altered (0 for 'index')
df_iqr = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

Interpretation:filtering out outlier

```python
df_iqr_.shape
```

Interpretation:shape of the IQR is (558, 12) which can be removed

```python
# check the shape of the data
df_iqr.shape
```

Interpretation:shape of data after removing outlier. rows have been reduced to 6827.

```python
df.info()
```

```python
# consider the variables which has outliers
variables = ['Engine Size(L)', 'Cylinders', 'Fuel Consumption City (L/100 km)', 'Fuel Consumption Hwy (L/100 km)', 
             'Fuel Consumption Comb (L/100 km)', 'Fuel Consumption Comb (mpg)']

# plot the boxplot for each variable
# subplots(): plot subplots
# figsize(): set the figure size
fig, ax = plot.subplots(2, 3, figsize=(15, 8))

# plot the boxplot using boxplot() from seaborn
# z: let the variable z define the boxplot
# x: data for which the boxplot is to be plotted
# orient: "h" specifies horizontal boxplot (for vertical boxplots use "v")
# whis: proportion of the IQR past the low and high quartiles to extend the plot whiskers
# ax: specifies the axes object to draw the plot onto
# set_xlabel(): set the x-axis label
# fontsize: set the font size of the x-axis label
for variable, subplot in zip(df_iqr[variables], ax.flatten()):
    z = sns.boxplot(x=df_iqr[variable], orient = "h",whis=1.5 , ax=subplot)
    z.set_xlabel(variable, fontsize = 20)
```

Interpretation:we can check if any more outliers are present. few more outliers are left in  Fuel Consumption City (L/100 km),Fuel Consumption Hwy (L/100 km),Fuel Consumption Comb (L/100 km) 

##### 3. Based on Z-score

```python
# import library scipy
import scipy
# from scipy import the module stats
from scipy import stats

# z-scores are defined for each observation in a variable
# compute the z-scores using the method zscore from the scipy library 
z_scores_comb_kms = scipy.stats.zscore(df_num["Fuel Consumption Comb (L/100 km)"])

# display the z-scores
z_scores_comb_kms
```

Interpretation:we can check z score to check what is the variation from top most and bottom most from mean. top most is -0.85 and bottommost is -0.09. these are the distance of points from the mean.this can also be used to check for outliers.

```python
# print the rows where z-score is less than -3
row_index_less = np.where(z_scores_comb_kms < -3)

# print the values
print(row_index_less)
```

```python
# print the rows where z-score is more than 3
row_index_more = np.where(z_scores_comb_kms > 3)

# print the values more than the 
print(row_index_more)
```

Interpretation:checking for outliers which are beyond 3. all the numbers beyond 3 are considered as outliers.

```python
# count of outliers in the variable representing profit
len(row_index_less[0]) + len(row_index_more[0])
```

```python
# filter out the outlier values
# ~ : selects all rows which do not satisfy the condition
df_zscore = df["Fuel Consumption Comb (L/100 km)"][~(( z_scores_comb_kms < -3) |(z_scores_comb_kms > 3))]
df_zscore.head(10)
```
